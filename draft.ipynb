## Market Watch

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://old.tsetmc.com/tsev2/data/MarketWatchInit.aspx?h=0&r=0"

response = requests.get(url)
```

```python
market_watch = BeautifulSoup(response.content).text
```

```python
len(market_watch.split("@"))
```

```python
market_watch.split("@")[0].split(",")
```

```python
filt = df[2].str.contains("اراد")
df.loc[filt]
```
